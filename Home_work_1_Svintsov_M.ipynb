## Task 1

Using the recurrence relation find out, that 
$$I_n(\alpha) = 1/n - \alpha I_{n-1}(\alpha)$$
$$I_0(\alpha) = ln|1+\alpha| - ln|\alpha|$$
Так же можем вычеслить формулу для обратной рекурсии.
$$I_{n-1}(\alpha) = \frac{1}{n\alpha} - \frac{1}{\alpha} I_{n}(\alpha)$$

Запищем обе функции

```python
def forward(a, n):
    I = np.log(1 + a) - np.log(a)
    answer = []
    for i in range(1, n + 1):
        answer.append(I)
        I = 1/i - a * I
    return answer
```

Сразу видно, что для а = 10 лучше работает обратная рекурсия, а для а = 0.1 прямая. Это можно объяснить взглянув на формулы. 
$$I_n(\alpha) = 1/n - \alpha I_{n-1}(\alpha)$$
В прямой рекурсии ошибка $\delta I$ накапливается из за слагаемого $$\alpha I_{n-1}(\alpha)$$ То есть в главном порядке по $alpha$ общая ошибка
$$\delta I_0 \thicksim \alpha^n \delta I$$ Тогда для $\alpha = 10$ ошибка накапливается очень быстро, и этот метод нам не подходит. Для $\alpha = 0.1$ все работает лучше, так как $\alpha = 0.1$ в положительной степение уменьшается.
В обратной рекурсии ошибка $$\delta I_0 \thicksim \frac{1}{\alpha^n} \delta I$$
Поэтому для обратной рекурсии получается посчитать ответ при $\alpha = 10$, а для $\alpha = 0,1$ работает плохо.

## Task 3

Заметим, что $n_i$ рекурентно задается с помощью формулы $$n_i = - n_{i-1} + 6 n_{i-2}$$
Запищем характеристическое уравнение $$\alpha^2 + \alpha - 6 = 0$$
$$\alpha_1 = -3, \ \alpha_2 = 2$$
Так же мы знаем, что $n_1 = -3$, а $n_2 = 2$

Тогда $$n_i = (-3)^i$$

```python
def recur(n):
    if n == 0:
        return 1
    if n == 1:
        return -3
    return -recur(n - 1) + 6 * recur(n - 2)
```

Проверим гипотезу на первых 20 значениях

```python
for i in range (1, 21):
    print(recur(i), 3**i)
```

Тогда итоговый ответ будет

```python
3**2020
```

```python
import math
def round_to_n(x, n):
    if x == 0:
        return x
    else:
        return round(x, -int(math.floor(math.log10(abs(x)))) + (n-1))
```

```python
res = 0
for k in range(1, 3001):
    res = round_to_n(res + 1/k**2, 4)
res
```

```python
res = 0
for k in range(3000, 0, -1):
    res = round_to_n(res + 1/k**2, 4)
res
```
