## Task 1

Using the recurrence relation find out, that $$I_n(\alpha) = 1/n - \alpha I_{n-1}(\alpha)$$
$$I_0(\alpha) = ln|1+\alpha| - ln|\alpha|$$

```python
def forward(a, n):
    I = np.log(1 + a) - np.log(a)
    answer = []
    for i in range(1, n + 1):
        answer.append(I)
        I = 1/i - a * I
    return answer
```

```python
def recur(n):
    if n == 0:
        return 1
    if n == 1:
        return -3
    return -recur(n - 1) + 6 * recur(n - 2)
```

```python
def solve_quad(b, c):
    x = (-b + (b**2 - 4 * c)**0.5) * 0.5
    y = (-b - (b**2 - 4 * c)**0.5) * 0.5
    return x, y
```

```python
from numpy import allclose
```

```python
variants = [{'b': 4.0, 'c': 3.0},
            {'b': 2.0, 'c': 1.0},
            {'b': 0.5, 'c': 4.0},
            {'b': 1e10, 'c': 3.0},
            {'b': -1e10, 'c': 4.0},
            {'b': 1e-10, 'c': 4.0},]
```

```python
for var in variants:
    print(solve_quad(**var))
```

```python
for var in variants:
    x1, x2 = solve_quad(**var)
    print(allclose(x1*x2, var['c']))
```
